```python
import yahoo_fin.stock_info as yf
import xlwings as xw
import pandas as pd
import datetime
import pandas_datareader.data as web
from dateutil.relativedelta import relativedelta
from multiprocessing.dummy import Pool as ThreadPool

#tickers = yf.tickers_sp500()
```

```python
df_list = pd.DataFrame()
list_wrong = []
i = 0

def incrs_max(ticker):
    
    global i 
  
    global list_wrong
    global df_list 
    i = i + 1
    names = locals()
    end = datetime.date.today()

    start = end - relativedelta(days=360)
    #df = web.get_data_yahoo(ticker,start,end)
    
    print(i,ticker)
    try:
            
        names['df' + str(ticker)] = web.get_data_yahoo(ticker,start,end)
        
        '''
        if names['df' + str(ticker)].iloc[-1]['Adj Close'] > names['df' + str(ticker)].iloc[-2]['Adj Close']:
            j = j + 1
        else:
            k = k + 1
        '''   
        names['start_price_360' + str(ticker)] = names['df' + str(ticker)].iloc[0]['Adj Close']
        names['start_price_180' + str(ticker)] = names['df' + str(ticker)].iloc[-180]['Adj Close']
        names['start_price_90' + str(ticker)] = names['df' + str(ticker)].iloc[-90]['Adj Close']
        names['start_price_30' + str(ticker)] = names['df' + str(ticker)].iloc[-30]['Adj Close']
        names['start_price_7' + str(ticker)] = names['df' + str(ticker)].iloc[-7]['Adj Close']
        names['start_price_2' + str(ticker)] = names['df' + str(ticker)].iloc[-2]['Adj Close']
        names['end_price' + str(ticker)] = names['df' + str(ticker)].iloc[-1]['Adj Close']
        
        names['increase_2' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_2' + str(ticker)] - 1
        names['increase_7' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_7' + str(ticker)] - 1
        names['increase_30' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_30' + str(ticker)] - 1
        names['increase_90' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_90' + str(ticker)] - 1
        names['increase_180' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_180' + str(ticker)] - 1
        names['increase_360' + str(ticker)] = names['end_price' + str(ticker)]/names['start_price_360' + str(ticker)] - 1
                               
                
        df_list = df_list.append(pd.DataFrame({'ticker':[ticker],\
                                        
                                         'start':names['df' + str(ticker)].index[0],\
                                         'end':names['df' + str(ticker)].index[-1],\
                                         #'start_price':names['start_price' + str(ticker)],\
                                         'end_price':names['end_price' + str(ticker)],\
                                         'start_price_7':names['start_price_7' + str(ticker)],\
                                         'start_price_30':names['start_price_30' + str(ticker)],\
                                         'start_price_90':names['start_price_90' + str(ticker)],\
                                         'start_price_180':names['start_price_180' + str(ticker)],\
                                         'start_price_360':names['start_price_360' + str(ticker)],\
                                               
                                         'increase_2':names['increase_2' + str(ticker)],\
                                         'increase_7':names['increase_7' + str(ticker)],\
                                         'increase_30':names['increase_30' + str(ticker)],\
                                         'increase_90':names['increase_90' + str(ticker)],\
                                         'increase_180':names['increase_180' + str(ticker)],\
                                         'increase_360':names['increase_360' + str(ticker)],

                                         }),ignore_index=True,sort=False) 
            
        
    except:
        print(ticker,'.....................wrong.....................')
        list_wrong.append(ticker)
    
```
